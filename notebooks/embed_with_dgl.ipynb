```python
# Node mapping with types
node_id_mapping = {}
node_type_data = {}

for _, row in nodes_df.iterrows():
    node_id = row['id']
    node_type = row['category']
    if node_type not in node_type_data:
        node_type_data[node_type] = []
    node_id_mapping[node_id] = len(node_type_data[node_type])
    node_type_data[node_type].append(node_id)
# print(len(node_mapping.keys()))
# print(len(node_id_mapping.keys()))
# print(node_types)
# print(node_type_data)

edges_df['subject_id'] = edges_df['subject'].map(lambda id: node_id_mapping[id])
edges_df['object_id'] = edges_df['object'].map(lambda id: node_id_mapping[id])
edges_df['subject_type'] = edges_df['subject'].map(lambda id: nodes_df[nodes_df['id'] == id]['category'].iloc[0])
edges_df['object_type'] = edges_df['object'].map(lambda id: nodes_df[nodes_df['id'] == id]['category'].iloc[0])

# Create a dictionary for edge data
graph_data = {}

for _, edge in edges_df.iterrows():
    edge_type = (edge['subject_type'], edge['predicate'], edge['object_type'])
    if edge_type not in graph_data:
        graph_data[edge_type] = ([], [])
    graph_data[edge_type][0].append(edge['subject_id'])  # Append subject node
    graph_data[edge_type][1].append(edge['object_id'])  # Append object node

# Convert lists to tensors for DGL
for edge_type, (src, dst) in graph_data.items():
    graph_data[edge_type] = (torch.tensor(src), torch.tensor(dst))

# Create the DGL Heterograph
g = dgl.heterograph(graph_data)
```

```python

# Convert to NetworkX graph for each edge type
graphs = []
for etype in g.canonical_etypes:
    src, dst = g.all_edges(etype=etype)
    nx_g = nx.DiGraph()  # Directed graph
    for s, d in zip(src.numpy(), dst.numpy()):
        src_node, dst_node = f'{etype[0]}_{s}', f'{etype[2]}_{d}'
        nx_g.add_edge(src_node, dst_node, label=etype[1])
    graphs.append(nx_g)

# Combine all the graphs
combined_nx_g = nx.compose_all(graphs)

# Prepare plot
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(combined_nx_g)  # Layout for the nodes

# Draw nodes and edges
nx.draw(combined_nx_g, pos, with_labels=True, node_size=30, node_color="lightblue", font_weight='bold')

# Draw edge labels
edge_labels = {(n1, n2): d['label'] for n1, n2, d in combined_nx_g.edges(data=True)}
nx.draw_networkx_edge_labels(combined_nx_g, pos, edge_labels=edge_labels)

plt.title('Heterograph Visualization')
plt.show()
```

```python
import optuna

# Set the logging level to INFO
# optuna.logging.set_verbosity(optuna.logging.I)

def make_rw(metapath: list, restart_prob: list):
    """Make random walk

    Args:
        restart_prob (_type_): _description_

    Returns:
        _type_: _description_
    """
    rw = dgl.sampling.random_walk(g, gene_nodes, return_eids=True, metapath=metapath, restart_prob=torch.FloatTensor(restart_prob))
    x = rw[0][:,1]
    return len(x[(x > 0) & (x < 100000)])


def objective(trial):
    # Define the range of parameters
    restart_prob1 = trial.suggest_float('restart_prob1', 0, 1)
    restart_prob2 = trial.suggest_float('restart_prob2', 0, 1)
    metapath =  trial.suggest_categorical('metapath', gene_metapaths)
    # metapath =  trial.suggest_categorical(gene_metapaths) #{param: trial.suggest_categorical(param, values) for param, values in gene_metapaths.items()}

    # Call the function with the suggested parameters
    result = make_rw(metapath=metapath, restart_prob=[restart_prob1, restart_prob2])

    # Optuna maximizes the return value of the objective function
    return result

# Create a study object and optimize the objective function
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1000, show_progress_bar=True)

# Print the best parameters
print(study.best_value)
print(study.best_params)
```

```python
from torch.optim import SparseAdam
from dgl.nn.pytorch import MetaPath2Vec
# from torch.utils.data import DataLoader
from dgl.dataloading import SAINTSampler, DataLoader

model = MetaPath2Vec(g, [('biolink:Gene', 'biolink:interacts_with', 'biolink:Gene')], window_size=1)
sampler = SAINTSampler(mode='node', budget=1000)

# Using pytorch's dataloader
# dataloader = DataLoader(torch.arange(g.num_nodes('biolink:Gene')), batch_size=128,
#                         shuffle=True, collate_fn=model.sample, drop_last=True, num_workers=0)
dataloader = DataLoader(g, indices=torch.arange(g.num_nodes('biolink:Gene')), 
                        graph_sampler=sampler, shuffle=True, drop_last=True, num_workers=0)
optimizer = SparseAdam(model.parameters(), lr=0.025)
```
